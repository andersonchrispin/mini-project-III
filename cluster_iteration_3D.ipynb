```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import plotly.express as px
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
```

```python
table = pd.read_csv('./Data/sample_for_cluster_algorithm.csv')
#table = table.loc[:, :'savings_balance' ]
table
```

```python
#reducing to 3 dimensions
pca_table_init = PCA(n_components=0.88, svd_solver="full")
pca_table_init.fit(table)
pca_table = pca_table_init.transform(table)

pca_table
```

```python
#initializing the centroids
np.random.seed(42)
centroids = [pca_table[0]]
k = 4

for _ in range(1, k):
    dist_sq = np.array([min([np.inner(c-x,c-x) for c in centroids]) for x in pca_table])
    probs = dist_sq/dist_sq.sum()
    cumulative_probs = probs.cumsum()
    r = np.random.rand()
        
    for j, p in enumerate(cumulative_probs):
        if r < p:
            i = j
            break
    centroids.append(pca_table[i])

centroids
```

```python
colors = {0:'y', 1:'darkgreen', 2:'b', 3:'r', 4:'m', 5:'k', 6:'c', 7:'g'}

for _ in range(10):
    #computing the clusters
    centroids_result = []
    for i in pca_table:
        diff = []
        for j in centroids:
            diff.append(np.linalg.norm(i - j))
        centroids_result.append(np.where(diff == min(diff))[0][0])
    
    #ploting iteration
    plt.figure(figsize=(12,5))
    plt3d = plt.axes(projection='3d')
    #plt3d = Axes3D(plt.figure(figsize=(12,5)))
    #clusters
    for i in range(len(pca_table)):
        plt3d.scatter(pca_table[i][0],
                        pca_table[i][1], 
                        pca_table[i][2],
                        c = colors[centroids_result[i]], 
                        #marker ="o",
                        #cmap = 'viridis',
                        #linewidths=0.5)
                        s = 1.5)
                        
    #centroids
    j = 0
    for i in range(k):
        plt3d.scatter(centroids[i][0],
                    centroids[i][1], 
                    centroids[i][2],
                    c = colors[j], 
                    #alpha=0.6,
                    s = 45)
        j += 1
    
    for i in range(k):
        plt3d.scatter(centroids[i][0],
                    centroids[i][1], 
                    centroids[i][2],
                    c = 'k', 
                    #alpha=0.6,
                    s = 7)
    plt3d.view_init(elev=45, azim=225)
    plt.show()
        
    #replacing centroids by new means
    centroids = []
    for i in set(centroids_result):
        temp = []
        c = 0
        for j in centroids_result:
            if (j == i):
                temp.append(pca_table[c]) 
            c += 1
        centroids.append(np.mean(temp, axis=0))
```
